```python
# Import setup
import pandas as pd
```

```python
# Create a list containing all the files for 2022 Jersey City data
files = ['data/JC_data/JC-202201-citibike-tripdata.csv',
         'data/JC_data/JC-202202-citibike-tripdata.csv',
         'data/JC_data/JC-202203-citibike-tripdata.csv',
         'data/JC_data/JC-202204-citibike-tripdata.csv',
         'data/JC_data/JC-202205-citibike-tripdata.csv',
         'data/JC_data/JC-202206-citibike-tripdata.csv',
         'data/JC_data/JC-202207-citibike-tripdata.csv',
         'data/JC_data/JC-202208-citibike-tripdata.csv',
         'data/JC_data/JC-202209-citibike-tripdata.csv',
         'data/JC_data/JC-202210-citibike-tripdata.csv',
         'data/JC_data/JC-202211-citibike-tripdata.csv',
         'data/JC_data/JC-202212-citibike-tripdata.csv']
```

```python
# Create an empty dataframe to store all data
citibike_df = pd.DataFrame()
```

```python
# Loop through the list of files and concatonate all data to the dataframe
for file in files:
    data = pd.read_csv(file) 
    citibike_df = pd.concat([citibike_df, data], axis=0)
```

```python
# Show the dataframe
citibike_df
```

```python
# Store the dataframe as CSV file
citibike_df.to_csv('data/2022jc_data.csv', index=False)
```
